**1** Using the data from the hourly time series at Deep Cove (`dc`), show that the canned `fft` routine returns the correct discrete Fourier Transform of a suitable subset of the data (your choice on what subset).  You can do this graphically, by comparing the amplitude and the phase of the Fourier components, or simply show that the resulting numbers are the same.  

Hints: 
  - ignore bad data (rather than dealing with it "properly").  Make sure the units of your FT are: $\mathrm{^oC\, s}$.
  - If you want to compute the DFT effeciently, form the matrix $F$ referred to in the notes, and plot some of the time series associated with the different entries and make sure they look like Sines and Cosines and that the edge conditions are what you want


````python
# this code taken from the answer key to assignment 3
def dft(x,dt,f):
    t = dt*np.arange(0,len(x))
    T,F = np.meshgrid(t,f)
    # make the fourier matrices
    E = np.exp(-1j*2*np.pi*T*F)
    X = dt*np.dot(E,x)
    return X
````

````python
N = 1000
C = 3600 #conversion factor
f = np.linspace(0,0.999,N)
x = dcg[1000:1000+N:]
X = dft(x,1,f) #calculate Fourier Transform Matrix 
Y = np.fft.fft(x) #canned routine

fig, ax = plt.subplots(1,1)
ax.loglog(f, np.real(X), label= 'X', alpha= 0.6, color= 'r')
ax.loglog(f, np.real(Y), label= 'Y', alpha= 0.6, color= 'b')
#ax[0].set_ylim([1e-1,1e5])
ax.set_title('Comparison of Fourier Transfom: Real')
ax.set_xlabel('F [Hz]')
ax.set_ylabel('Amplitude')
ax.legend(loc=3)

fig, bx= plt.subplots(1,1)
bx.loglog(f, np.imag(X), label= 'X', alpha= 0.6, color= 'r')
bx.loglog(f, np.imag(Y), label= 'Y', alpha= 0.6, color= 'b')
bx.set_title('Comparison of Fourier Transfom: Imaginary')
bx.set_xlabel('F [Hz]')
bx.set_ylabel('Phase')
bx.legend(loc=3)
plt.tight_layout()
````

There's one aberrant data point in the canned script, otherwise the data set for both real and imaginary portions of the Fourier Transform are identical.

**1** Write a short function to create a "red" time series by taking the cumulative sum of Normally distributed random data, and plot one realization.  The number of data points $N$ can be relatively small (say 1000).

````python
def red(n):
    R = np.zeros(n)
    r = np.random.randn(n)
    for i in range(0,n):
        R[i] = np.sum(r[0:i])
    return R

rr = red(N)
fig, cx=plt.subplots()
cx.plot(rr, color = 'r')
cx.set_title('"Red" Function')
cx.set_ylabel('Cumulative Sum')
cx.set_xlabel('Random Number Index')
plt.tight_layout()
````

**2** Write a second function to return the estimate of the power spectrum of $x$.  For posterity sake, include a value of the sampling frequency, and return the proper frequencies for the estimate.  Plot the resulting spectrum.

````python
def gxx(v):
    n = np.size(v)
    W = np.fft.fft(v)
    g = np.real(W[:n/2]*np.conj(W[:n/2])*2/n)
    return g

N = 1000
f = np.linspace(0,1,N/2)
Y = np.fft.fft(rr) #canned routine

fig, dx = plt.subplots(1,1)
dx.loglog(f, gxx(rr), label= 'R', color= 'r')
dx.set_title('Power Spectrum "Red" Function')
dx.set_xlabel('F [Hz]')
dx.set_ylabel(r'$G_{xx}$')
plt.tight_layout()
````
