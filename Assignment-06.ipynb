### Error bounds on coherence estimates

<div style='background:#F0F0F0'>  The 95% confidence intervals for the coherence squared $\gamma^2_{xy}(f)$ between two variables $x(t)$ and $y(t)$ are approximately:
$$ \gamma^2_{xy} \pm \frac{\sqrt{2}\left(1+\gamma^2_{xy}\right)}{\left|\gamma_{xy} \right|\sqrt{n_d}}$$
where $n_d$ is the number of blocks used to calculate the coherence.  

You **may** use `matplotlib.mlab.csd` for the following question, but you had best use it correctly!
</div>

<div style='background:#F0F0F0'>**1** Make an input process $x(t)$ as random Gaussian noise, and the output process the sum of a noise source and a running average of the input: $y(t)=n(t)+\frac{1}{b}\int_{t-b}^t x(t')dt'$ and plot (Zoom so we can see whats going on, and label your plot).  Make sure you generate a large time series.</div> 

```python
t = 32768
b = 64
n1 = np.random.randn(t)

x = np.random.randn(t+b-1)

x2 = np.convolve(x,ones(b),mode='valid')/b

x = x[:t]
y = n1 + x2

fig, ax=plt.subplots(2,1, figsize=(8,8))
ax[0].plot(x, alpha=0.6, label = 'x')
ax[0].plot(y, alpha=0.6, label = 'y')
ax[1].plot(x[1:50], label = 'x')
ax[1].plot(y[1:50], label = 'y')

ax[0].set_title('Random Noise and Dependent Function Y')
ax[0].set_ylabel('Amplitude')
ax[0].set_xlabel('Time')
ax[0].legend(loc=1)

ax[1].set_title('Close-Up: Random Noise')
ax[1].set_ylabel('Amplitude')
ax[1].set_xlabel('Time')
ax[1].legend(loc=1)
plt.tight_layout()
```

<div style='background:#F0F0F0'>**2** Compute the coherence squared of the input and the output.  Make a plot with the spectra in the upper panel and the coherence squared in the middle panel and the phase in the bottom.

You will need to choose a value for your fft length.  Choose a relative low one to get a smooth plot.
</div>

```python
coh, f = mlab.cohere(x, y, NFFT= b, Fs=1, noverlap=b/2)
xsd, fx = mlab.csd(x, y, NFFT= b, Fs=1, noverlap=b/2)
px, xf = mlab.psd(x, NFFT=b, Fs=1, noverlap=b/2)
py, yf = mlab.psd(y, NFFT=b, Fs=1, noverlap=b/2)

print size (py)
fig, bx=plt.subplots(3,1, figsize=(8,8))
bx[0].semilogx(xf[1:len(px)-1], px[1:len(px)-1], alpha=0.6, label = 'x')
bx[0].semilogx(xf[1:len(py)-1], py[1:len(py)-1], alpha=0.6, label = 'y')
bx[1].semilogx(f[1:len(coh)-1], coh[1:len(coh)-1], color='r')
bx[2].semilogx(fx, -angle(xsd)/pi)
bx[0].set_title('Power Specturm, Coherence and Phase: Noise and Dependent Function Y')
bx[0].set_ylabel('Power Spectrum')
bx[1].set_ylabel('Coherence')
bx[2].set_ylabel('Phase')
bx[2].set_xlabel('Frequency')
bx[0].legend(loc=1)
plt.tight_layout()
```

**COMMENT**:
> The two signals appear to have some commonality. Coherence is fairly smooth, with peaks occuring where the two functions line up. Phase seems to be pretty random.

<div style='background:#F0F0F0'>**3** Now repeat the above for three different values of NFFT, each 2 to 5 times as large as the next.  Plot the resulting coherence squared functions in separate panels.  Add the confidence intervals to your plot and comment on which peaks are significant.
</div>

```python
nfft = [128, 256, 1024]
fig, cx=plt.subplots(3,1, figsize=(8,8))
for i, k in enumerate(nfft):
    coh, f = mlab.cohere(x, y, NFFT=k, Fs=1, noverlap=k/2)
    xsd, fx = mlab.csd(x, y, NFFT=k, Fs=1, noverlap=k/2)
    px, xf = mlab.psd(x, NFFT=k, Fs=1)
    py, yf = mlab.psd(y, NFFT=k, Fs=1)

    cx[i].loglog(xf, px, alpha=0.6, label = 'PSD: x')
    cx[i].loglog(xf, py, alpha=0.6, label = 'PSD: y')
    cx[i].loglog(f, coh, color='r', label = 'Coherence')
    
cx[0].set_title('Coherence and Power Spectrum for Increasing Values of NFFT ')
cx[1].set_ylabel('Coherence')
cx[2].set_xlabel('Frequency')
cx[0].legend(loc=2)
plt.tight_layout()
```

**COMMENT**:
> As NFFT increases, more variation appears between the power spectrum of the two signals, and thus coherence also becomes more and more erratic.

## Q2: Shear turbulence data

```python
time=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/time.txt')
ac=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/ac.txt')
shear=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/s.txt')
```


<div style='background:#F0F0F0'>Download the data abaove. `time` is a measurement time in seconds. `shear` is a little less than 300 s of data from a turbulence shear probe. Units are $s^{-1}$. `ac` is a suitably scaled acceleration of the vehicle containing the shear probes.

**1** Plot the shear time series and the accelerometer time series.  Comment on any similarities and differences.  Comment on the stationarity of the time series.</div>

```python
fig, dx=plt.subplots(1,1, figsize=(8,6))
dx.plot(time, ac, alpha=0.6, label = 'ac')
dx.plot(time, shear, alpha=0.6, label = 'shear')

dx.set_title('Shear Probe Data Overlaid with Vehicle Acceleration')
dx.set_ylabel('Amplitude')
dx.set_xlabel('Frequency')
dx.legend(loc=1)
plt.tight_layout()
```

**COMMENT**:
> Acceleration values are consistently lower and more compact than the shear values. Both become more erratic in sections of data near the beginning and at the very end. Shear values also appear to be almost symmetric about 0. Data is for both oscillates dramaticallly.

<div style='background:#F0F0F0'> **2** Compute the power spectra of the two signals and the coherence squared (you could do the phase too, but it is not too thrilling).  Comment on the spectral similarities and the coherence. </div> 

```python
coh, f = mlab.cohere(ac, shear, Fs=1, noverlap=128)
pac, acf = mlab.psd(ac, Fs=1, noverlap=128)
psh, shf = mlab.psd(shear, Fs=1)

fig, ex=plt.subplots(1,1, figsize=(8,8))
ex.loglog(acf, pac, alpha=0.6, label = 'Ac')
ex.loglog(shf, psh, alpha=0.6, label = 'Shear')
ex.loglog(f, coh, color='r', label= 'Coherence')

ex.set_title('Power Spectrum and Coherence: Ac and Shear Data')
ex.set_ylabel('Spectrum')
ex.set_xlabel('Frequency')
ex.legend(loc=1)
plt.tight_layout()
```

**COMMENT HERE:**
> The does appear to be some similarity between the shear power spectrum and the ac power spectrum, with several of the peaks aligning. Coherence betweeen the two signals reflects this.

<div style='background:#F0F0F0'> **3** Compute a spectra that has the contamination from the vibration removed using the linear technique described in the notes. Comment on how well your correction worked.</div>

```python
Gnn =  psh - coh*psh

fig, fx=plt.subplots(1,1, figsize=(6,6))
fx.loglog(f, Gnn, label='Recovered Noise')
fx.legend()

fx.set_title('')
fx.set_ylabel('')
fx.set_xlabel('Frequency')
fx.legend(loc=1)
plt.tight_layout()
```

**COMMENT:**
> The recovered data is very similar to that of the initial shear data, including the sizeable peak near the end of the series.

<div style='background:#F0F0F0'>**4** Chose two separate periods of the time series by eye, one that is "turbulent" and one that is "quiet", and plot the raw and corrected shear spectra.  Comment on the differences and the effect of the correction on each spectral estimate.    </div>

```python
fig, gx=plt.subplots(1,2, figsize=(10,6))
gx[0].loglog(Gnn[75:90], label='Recovered')
gx[0].loglog(psh[75:90], color='r', label='Raw')
gx[1].loglog(Gnn[35:50], label='Recovered')
gx[1].loglog(psh[35:50], color='r', label='Raw')

gx[0].set_title('Raw and Recovered Data: Turbulent')
gx[1].set_title(':Quiet')

gx[0].set_ylabel('Power Spectrum')
gx[0].set_xlabel('Frequency')
gx[1].set_xlabel('Frequency')

gx[0].legend(loc=1)
gx[1].legend(loc=1)

plt.tight_layout()
```

**COMMENT:**
> In portions of the data where it was relatively calm, the recovered data is almost identical to the raw. The peak recovered from the more erratic data is slightly lower than the raw data indicates, but some reduction in magnitude is to be expected with the random element of the interference.
