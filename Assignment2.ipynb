```python
sample = 200000
a1 = 1.0
a2 = 10.0
x1 = np.random.randn(sample)

b1 = np.zeros(40)
s1 = np.zeros(40)
r1 = np.zeros(40)
sd1 = np.zeros(40)

b2 = np.zeros(40)
s2 = np.zeros(40)
r2 = np.zeros(40)
sd2 = np.zeros(40)

space = np.logspace(-2, 2, 40, 10)
for i in range(0,40):
    n1 = np.random.randn(sample)*space[i]
    n2 = np.random.randn(sample)*space[i]
    
    y1 = a1*x1 + n1
    y2 = a2*x1 + n2
    
    b1[i] = np.sum((y1 - np.mean(y1))*x1)/np.sum((x1 - np.mean(x1))*x1)
    b2[i] = np.sum((y2 - np.mean(y2))*x1)/np.sum((x1 - np.mean(x1))*x1)
    aa1 = np.mean(y1) - np.mean(x1)*b1[i]
    aa2 = np.mean(y2) - np.mean(x1)*b2[i]
    
    yy1 = b1[i]*x1 + aa1
    yy2 = b2[i]*x1 + aa2
    
    s1[i] = (np.sum((y1 - yy1)**2)/(sample - 2))**0.5
    s2[i] = (np.sum((y2 - yy2)**2)/(sample - 2))**0.5
       
    r1[i] = (np.sum(x1*y1)/np.std(x1,ddof=1)/np.std(y1,ddof=1)/(sample-1))
    r2[i] = (np.sum(x1*y2)/np.std(x1,ddof=1)/np.std(y2,ddof=1)/(sample-1))
    
    sd1[i] = np.std(x1)*aa1/sample
    sd2[i] = np.std(x1)*aa2/sample

fig, ax1 = plt.subplots(1,2)

ax1[0].hist(b1)
ax1[0].set_xlabel('Slope')
ax1[0].set_ylabel('Occurance')
ax1[0].set_title('Slope Distrubution')

ax1[1].hist(s1)
ax1[1].set_xlabel('Sample Standard Deviation')
ax1[1].set_ylabel('Occurance')
ax1[1].set_title('SSD Distrubution')

plt.tight_layout()
```

```python
fig, ax1b = plt.subplots(1,2)

ax1b[0].loglog(r1, sd1)
ax1b[0].set_xlabel('Correlation Coefficient')
ax1b[0].set_ylabel('Normalized Noise')
ax1b[0].set_title('Noise vs Correlation Coefficient')

ax1b[1].loglog(r2, sd2)
ax1b[1].set_xlabel('Correlation Coefficient')
ax1b[1].set_ylabel('Normalized Noise')
ax1b[1].set_title('Noise vs Correlation Coefficient')

plt.tight_layout()
```

Comments:
---------
The slope distribution is has the greatest number of occurances near 1, with a few random outliers. This is reasonable given it's a random sample. The standard deviations similarly suggest tight results around the mean with decreasing deviation.

Experimentation with amplitude and noise revealed that the correlation coefficient goes up proportionally to amplitude and down with respect to noise. Very large values of 'a' with comparably small 'n' produce the highest correlation.

```python
# Procedure accepts points, noise and amplitude and calculates the slope and sample standard deviation of a random data set
# using those parameters.

# Returns slope and ssd.
def slope(points, noise, amp):

    ns = np.random.randn(points)*noise

    xx = np.random.randn(points)
    yy = amp*xx + ns

    bb = np.sum((yy - np.mean(yy))*xx)/np.sum((xx - np.mean(xx))*xx)
    aa = np.mean(yy) - np.mean(xx)*bb
    yyy = bb*xx + aa
    
    ssd = (np.sum((yy - yyy)**2)/(points - 2))**0.5
    return (bb, ssd)    
```

```python
test = 10000
data = np.zeros((2, test)) # data is an array of (slope, ssd) 

for i in range(0, test):
    data[0][i], data[1][i] = slope(10, 1, 1)


s1 = np.random.standard_t(test-2, size=10000) + 1

fig, ax2 = plt.subplots(1,2)
ax2[0].hist(data[0])
ax2[0].set_xlabel('Slope')
ax2[0].set_ylabel('Occurance')
ax2[0].set_title('Slope Distrubution')

ax2[0].hist(s1, alpha = 0.5)

ax2[1].hist(data[1])
ax2[1].set_xlabel('Sample Standard Deviation')
ax2[1].set_ylabel('Occurance')
ax2[1].set_title('SSD Distrubution')
plt.tight_layout()
```

```python
test = 10000
data2 = np.zeros((2, test)) # data is an array of (slope, ssd) 

for i in range(0, test):
    data2[0][i], data2[1][i] = slope(5, 3, 1)

s2 = np.random.standard_t(test-2, size=10000) + 1

fig, ax3 = plt.subplots(1,2)
ax3[0].hist(data2[0])
ax3[0].set_xlabel('Slope')
ax3[0].set_ylabel('Occurance')
ax3[0].set_title('Slope Distrubution')
 
ax3[0].hist(s2, alpha=0.5)

ax3[1].hist(data2[1])
ax3[1].set_xlabel('Sample Standard Deviation')
ax3[1].set_ylabel('Occurance')
ax3[1].set_title('SSD Distrubution')
plt.tight_layout()
```

Comments:
---------

The two histograms for the first sample are symmmetric with a mean ~= 1.0. The student_t distribution lines up well with the histogram for the slope of the sample.

Changing the values (I reduced the amplitude and increased the noise), resulted in far less consistency in the results, which is to be expected given the correlation coefficient should also go down given those changes. The sample standard deviation is no longer symmetric.

```python
# Load the data:

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb = hourdata[1,:]

good = np.isfinite(dc + jb)

dcg = dc[good] 
dcg = dcg - np.mean(dcg)
jbg = jb[good]
jbg = jbg - np.mean(jbg)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
im = ax.scatter(dcg, jbg, cmap=plt.cm.jet)

r = (np.sum((dcg - np.mean(dcg))*(jbg - np.mean(jbg)))/np.std(dcg,ddof=1)/np.std(jbg,ddof=1)/(np.size(good)-1))
print 'Correlation coefficient is %f'%(r)

b = np.sum((jbg - np.mean(jbg))*dcg)/np.sum((dcg - np.mean(dcg))*dcg)
a = mean(jbg) - b*mean(dcg)

y = b*dcg + a
ax.plot(dcg, y, color = 'g')

s = sqrt(np.sum((jbg - y)**2)/(np.size(good)-2)) #error calculation
print 'Mean error %f'%(s) 

import scipy.stats as stats
b_dn, b_up = stats.t.interval(0.95, df=(np.size(good)-2), scale=s)
bb = np.array([b_dn, b_up])
bb = bb/np.sqrt(np.sum((jbg - np.mean(jbg))**2))
print 'Slope b is between %f and %f with 95%% probability'%(b+bb[0],b+bb[1])

x = np.sort(jbg)
y_high = (b+bb[1])*(x - np.mean(jbg)) + np.mean(y)
y_low = (b+bb[0])*(x - np.mean(jbg)) + np.mean(y)
ax.fill_between(x, y_high, y_low, color = 'r', alpha=0.6)
ax.set_xlabel('Deep Cove Temperature')
ax.set_ylabel('James Bay Temperature')
ax.set_title('Correlation of James Bay vs Deep Cove Temperatures');
```
