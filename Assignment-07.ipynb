### **Q1** Frequency-domain filtering

<div style='background:#F0F0F0'>  Consider a time series with a sharp jump in the middle as in the notes:
</div>

```python
%matplotlib inline
import numpy as np
import math as math
import numpy.random as random
import scipy.stats as stats
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from pylab import *

N = 1001
dt = 1.
t = np.arange(0,N*dt,dt)

x = np.cumsum(randn(N)) # red noise
f0 = 0.03 # Hz
x = x+ 40.*sin(f0*2*pi*t) # + sine wave
x = x-np.mean(x)
noise = 60.*randn(N)
xn = x+noise
xn[N/2:] = xn[N/2:]-2000.
# remove the mean!!
xn = xn-np.mean(xn)
```

<div style='background:#F0F0F0'>**1** Implement a low-pass filter of the data in the frequency domain, with a cutoff frequency of $f_c=0.06\ Hz$.</div>
   
  1. plot the signal and the smoothed signal (suggest zooming in on the step).
  2. plot the spectral response $|H(f)|$
  3. plot the phase response $\mathrm{angle}(H(f))$.

<div style='background:#F0F0F0'>
Be sure to comment on how good your filter is.
</div>

```python
fc = 0.06
n = 4 

XN = np.fft.fft(xn)
ff = np.linspace(0,1./dt,N)
XN[(ff > fc ) & (ff < (1./dt - fc))] = 1.e-15
y = real(np.fft.ifft(XN))+1.e-17*np.random.randn(N)

psxn, f = mlab.psd(xn, Fs= 1./dt)
psy, f = mlab.psd(y, Fs= 1./dt)
H = (psy/psxn)**0.5

fig, ax=subplots(4,1,figsize=(6,16))

ax[0].plot(t, x, 'b', label='signal')
ax[0].plot(t, xn, 'k', label='signal+noise')
ax[0].plot(t, y, 'r', label='filtered')

ax[1].plot(t, xn, 'k', label='signal+noise')
ax[1].plot(t, y, 'r', label='filtered')

ax[2].loglog(f,H)

pxny, f = mlab.csd(xn, y, Fs=1./dt)

ax[3].semilogx(f, angle(pxny)/np.pi, 'k')
ax[0].set_title('Full Signal Plus Noise and Filter');ax[1].set_title('Zoom: Step')
ax[2].set_title('Spectral Response in Frequency Domain');ax[3].set_title('Phase Shift')
ax[0].legend(fontsize='small');ax[1].legend(fontsize='small')
ax[0].set_xlabel('Time');ax[1].set_xlabel('Time');ax[2].set_xlabel('f [Hz]');ax[3].set_xlabel('f [Hz]')
ax[0].set_ylabel('Amplitude');ax[1].set_ylabel('Amplitude');ax[2].set_ylabel('Sprectral Response H');ax[3].set_ylabel('phase / $\pi$')

ax[1].set_xlim(400,600)

plt.tight_layout()
```

COMMENT HERE:

> Filter has a large ripple near the step, otherwise you can see the general form of the sin wave in the filtered signal, with variation from the random noise.

<div style='background:#F0F0F0'>**2** Now repeat the above but apply a Hanning window to the frequency window *before* you apply it to the FFT of $x_n$.  Comment on how well the filter is doing when you do this.  (Its helpful to keep the previous result up)
</div>

```python
wlo = np.hanning(2*fc*N)[fc*N:2*fc*N]
whi = np.hanning(2*fc*N)[0:fc*N]
ff[0:fc*N] = wlo
ff[1-(fc+N**-1)*N:N] = whi
HX = XN*ff

yy = real(np.fft.ifft(HX))+1.e-17*np.random.randn(N)

psyy, f = mlab.psd(yy, Fs= 1./dt)
H = (psyy/psxn)**0.5

fig, ax=subplots(4,1,figsize=(6,16))
ax[0].plot(t, x, 'b', label='signal+noise')
ax[0].plot(t, xn, 'k', label='signal+noise')
ax[0].plot(t, yy, 'r', label='filtered')

ax[1].plot(t, xn, 'k', label='signal+noise')
ax[1].plot(t, yy, 'r', label='filtered')

ax[2].loglog(f,H)

pxny, f = mlab.csd(xn, yy, Fs=1./dt)

ax[3].semilogx(f, angle(pxny)/np.pi, 'k')

ax[0].set_title('Full Signal Plus Noise and Filter');ax[1].set_title('Zoom: Step')
ax[2].set_title('Spectral Response in Frequency Domain');ax[3].set_title('Phase Shift')
ax[0].legend(fontsize='small');ax[1].legend(fontsize='small')
ax[0].set_xlabel('Time');ax[1].set_xlabel('Time');ax[2].set_xlabel('f [Hz]');ax[3].set_xlabel('f [Hz]')
ax[0].set_ylabel('Amplitude');ax[1].set_ylabel('Amplitude');ax[2].set_ylabel('Sprectral Response H');ax[3].set_ylabel('phase / $\pi$')

ax[1].set_xlim(400,600)

plt.tight_layout()
```

COMMENT

<div style='background:#F0F0F0'>**3** So the Hanning window falls off a *lot* earlier and more slowly than a step function, so adjust the cutoff frequency to get a reasonable filter that preserves the amplitude of the sine wave at $f=0.03\ \mathrm{Hz}$.  Comment on the pros and cons of this approach.  
</div>

```python
fo = 0.03
fc = 0.06
ff = 1.e-15*np.ones(N)
wlo = np.hanning(2*fc*N)[fc*N:2*fc*N]
whi = np.hanning(2*fc*N)[0:fc*N]


ff[fo*N:(fo+fc)*N] = wlo
ff[1-(fc+fo)*N:1-N*fo] = whi

ff[1-N*fo:N] = 1
ff[0:fo*N] = 1

HX = XN*ff

yy = real(np.fft.ifft(HX))+1.e-17*np.random.randn(N)

psyy, f = mlab.psd(yy, Fs= 1./dt)
H = (psyy/psxn)**0.5

fig, ax=subplots(3,1,figsize=(6,12))
ax[0].plot(t, xn, 'k', label='signal+noise')
ax[0].plot(t, y, 'b', label='original filter')
ax[0].plot(t, yy, 'r', label='filtered')

ax[1].loglog(f,H)
ax[1].set_ylim()

pxny, f = mlab.csd(xn, yy, Fs=1./dt)

ax[2].semilogx(f, angle(pxny)/np.pi, 'k')
ax[2].axvline(x=0.03)


ax[0].set_title('Zoom: Step');ax[1].set_title('Spectral Response in Frequency Domain');ax[2].set_title('Phase Shift')
ax[0].legend(fontsize='small')
ax[0].set_xlabel('Time');ax[1].set_xlabel('f [Hz]');ax[2].set_xlabel('f [Hz]')
ax[0].set_ylabel('Amplitude');ax[1].set_ylabel('Sprectral Response H');ax[2].set_ylabel('phase / $\pi$')

ax[1].set_xlim(400,600)

plt.tight_layout()
```

## Q2: Implementing a bandpass filter

<div style='background:#F0F0F0'>**1** Create a red-noise process with 20000 data points, $dt=1\ s$.  Using a **truncated Dirichlet Kernel**, create a **bandpass** filter centered at $f_c=0.1\ Hz$ that has a passband that is approximately 0.1 Hz wide. Choose a filter length that is long enough to accomodate 0.1 Hz. 
</div>

   1. Plot the unshifted and shifted filters versus $k$, the time index.
   2. Plot the spectra, the response function $H^2(f)$ and the phase response
   3. Comment on your bandpass and if it indeed passed the right frequencies.

HINT: when you make the Dirichlet Kernel filter make sure that the $k=0$ case is treated "correctly" because $sin(0)=0$ and you will get a discontinuity.  Calculate what $h_0$ should be from [L'Hôpital's_rule](http://en.wikipedia.org/wiki/L'Hôpital's_rule).

d/dt [1/N\*sin(pi\*k\*M/N)] / d/dt [sin(pi\*k/N)]
=> M/N\*[cos(pi\*k\*M/N) / cos(pi\*k/N)]
=> M/N for k=0

```python
N = 20000
dt = 1.
t = np.arange(0,N*dt,dt)
k = np.array(range(-N/2+1,N/2))

xx = np.cumsum(randn(N)) # red noise
xx = xx - np.mean(xx)

fl = 20
fc = 0.1
K = N/2
M = fc*2*N+1

k[N/2-1] = 1
h = (1./N)*np.sin(np.pi*k*M/N)/np.sin(np.pi*k/N)
h[N/2-1] = M/N
k[N/2-1] = 0
h[abs(k)>(M-1)/2]=0
H = np.ones(size(k))

H[(np.abs(k*1.) > M/2)]=0.
h1 = np.abs(np.fft.fftshift(np.fft.ifft(np.fft.fftshift(H))))
h1 = h1[abs(k)<fl+1]

Hns = N*(ifft(h))
Hs = N*fftshift(ifft(h))

px, f = mlab.psd(x, Fs=dt)
f = abs(k*1./len(k))

fig,ax=plt.subplots(4,1,figsize=(6,16))
ax[0].plot(k, h, label = 'Unshifted $h_k$')
ax[0].plot(k[abs(k)<fl+1],h1,'r', label= 'Shifted $h_k$:Filter Length 20')

y = np.convolve(xx, h1,mode='same')

ax[1].plot(xx, label='Raw')
ax[1].plot(y, label='Filtered',color='r')

ax[2].semilogx(f, angle(Hs)/pi)
ax[3].loglog(f, abs(Hs**2))

ax[0].set_title('Filter in Time Domain');ax[1].set_title('Filtered Data')
ax[2].set_title('Phase Angle');ax[3].set_title('Response Function')
ax[0].legend(fontsize='small');ax[1].legend(fontsize='small')
ax[0].set_xlabel('k');ax[1].set_xlabel('t [s]');ax[2].set_xlabel('f [Hz]');ax[3].set_xlabel('f [Hz]')
ax[0].set_ylabel('$h_k$');ax[1].set_ylabel('x [V]');ax[2].set_ylabel('Phase / $\pi$')
ax[3].set_ylabel('Sprectral Response $H^2$')

ax[0].set_xlim(-50,50)
ax[1].set_xlim()

plt.tight_layout()
```

COMMENT
> Compairing the raw and filtered data shows that the values near zero agree, whereas the larger the magnitude of the raw data, the more is being removed by the filter. The spectral response function shows everything with frequency less than 0.1 is getting chopped off.

```python
dt = 1
t = np.arange(0,len(dc)*dt,dt)

Rp = 0.5 # Ripple in dB (passband)
Rs = 20  # Attenuation
Wp = 0.017 # Cuttoff frequency (hourly)
Ws = 0.034 # Stopband frequency

N, Wn = signal.ellipord(Ws, Wp, Rp, Rs)
B, A = signal.ellip(N, Rp, Rs, Wn, 'low')

dcl = signal.lfilter(B, A, dc)

Wp = 0.0007 # Cuttoff frequency (daily)
Ws = 0.0014 # Stopband frequency
N, Wn = signal.ellipord(Ws, Wp, Rp, Rs)
B, A = signal.ellip(N, Rp, Rs, Wn, 'high')

dch = signal.lfilter(B, A, dcl)

pdc, f = mlab.psd(dc, NFFT=(60*24*14), Fs=1)
pdy, f = mlab.psd(dch, NFFT=(60*24*14), Fs=1)

fig, ax=plt.subplots(1, 1, figsize=(8,4))

ax.loglog(f, pdc)
ax.loglog(f, pdy)
ax.set_title('Power Spectrum: Raw and Filtered Data')
ax.set_xlabel('f [hz]')
ax.set_ylabel('$G_xx$')
plt.tight_layout()
```
