```python
%matplotlib inline
import numpy as np
import math as math
from scipy.stats import t
import urllib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from pylab import *
```

#Question 1:

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb = hourdata[1,:]

good = np.isfinite(dc + jb)

dcg = dc[good] 
dcg = dcg - np.mean(dcg)
jbg = jb[good]
jbg = jbg - np.mean(jbg)
```

```python
lags = arange(0,720)
cxx = 0.*lags;
for i, tau in enumerate (lags):
    if (tau == 0):
        cxx[i] = np.mean(dcg[:]*dcg[:]/var(dcg))
    else:
        cxx[i] = np.mean(dcg[:-tau]*dcg[tau:]/var(dcg))
        
fig, ax = plt.subplots(1,1)
ax.plot(lags, cxx)
ax.set_title('Lag Correlation Coefficient: 30 days')
ax.set_xlabel(r'$\tau [hr]$')
ax.set_ylabel('$C_{xx}$')
plt.tight_layout()
```

> You can basically see the day-to-day temperature oscillating up and down with respect to the average temperature recorded for the full data set. Correlation drops off slowly from the beginning to the end of the month.

```python
# taking winter and summer values
dcw = dc[0:2160:]
dcs = dc[4320:6480:]
g2 = np.isfinite(dcw + dcs)

dcw = dcw[g2]
dcs = dcs[g2]
dcw = dcw - np.mean(dcw)
dcs = dcs - np.mean(dcs)
```

```python
lags = arange(0,720)
cww = 0.*lags;
css = 0.*lags;
for i, tau in enumerate (lags):
    if (tau == 0):
        cww[i] = np.mean(dcw[:]*dcw[:])/var(dcw)
        css[i] = np.mean(dcs[:]*dcs[:])/var(dcs)
    else:
        cww[i] = np.mean(dcw[:-tau]*dcw[tau:])/var(dcw)
        css[i] = np.mean(dcs[:-tau]*dcs[tau:])/var(dcs)
        
fig, ax = plt.subplots(1,1)
ax.set_title('Lag Correlation Coefficient: Seasonal')
ax.plot(lags, cww, label = 'Winter')
ax.plot(lags, css, label = 'Summer')
ax.set_xlabel(r'$\tau [hr]$')
ax.set_ylabel('$C_{xx}$')
ax.legend(loc=1);
```

>Correlation oscillates more for the seasonal data than in the one month sample. One can see that the temperature in the winter changes less from night to day. Presumably the up and down nature of the correlation reflects weather trends during the season.

```python
n = [500 , 5000, 50000, 500000]
fig, ax = plt.subplots(1,1)
b = 50

lags = arange(0,100)
rxx = np.zeros((np.size(n),np.size(lags)))

for k, j in enumerate(n):
    x1 = np.zeros(j-b)
    y = np.random.randn(j)
    for i in range(0, j-b):
        x1[i] = np.sum(y[i:i+b])

    for i, tau in enumerate (lags):
        if (tau == 0):
            rxx[k][i] = np.mean(x1[:]*x1[:])
        else:
            rxx[k][i] = np.mean(x1[:-tau]*x1[tau:])

    ax.plot(lags, rxx[k], label = j)
    ax.set_title('Lag Correlation for Random Noise')
    ax.set_xlabel('Time Lag')
    ax.set_ylabel('$R_{xx}$')
    ax.legend(loc=1);
```

> Since the data set is random, the only data which correlates is the overlap between the time lagged data sets. One can see the triangular shape of the correlation as the time lag overlaps less and less. One can also see that just due to random variation, there is some correlation for the smaller data sets even when the time lag does not overlap. This is eliminated for suitably high data sets.
