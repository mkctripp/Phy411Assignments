### **Q1** Sampling theorem and aliasing

<div style='background:#F0F0F0'>**1** Create a band-limited white noise timeseries with $N=1000$ entries, $dt=0.1 s$, but with all frequencies greater than 0.5 Hz filtered.  Plot the power spectrum to make sure you got it correctly...
</div>

```python
N=1e3
dt = 1e-1
x0 = np.random.randn(N)
t = np.arange(0,N*dt,dt)

#Bandwidth = 2 Hz
B = 0.5
# filter to make Bandlimited signal
fp = B*0.85
fs = B*.95
fNyq=1/2./dt
n, fn = signal.ellipord(fp/fNyq, fs/fNyq,.1, 60.)
b, a = signal.ellip(n,.1,60.,fn)
xBL = signal.filtfilt(b,a,x0)

nfft=1024
pxx,f=mlab.psd(x0,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=window_hanning)
pxBL,f=mlab.psd(xBL,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=window_hanning)

fig,ax=plt.subplots(2,1,figsize=(8,8))

ax[0].plot(t,x0,color='0.5',label='full-spectrum signal')
ax[0].plot(t,xBL,color='k',label='band-limited signal')

ax[1].loglog(f,pxx,color='0.5')
ax[1].loglog(f,pxBL,color='k')

ax[0].set_xlim([0,30]);ax[0].set_xlabel('time [s]');ax[0].set_ylabel('x [V]');ax[0].legend(fontsize='small',loc=0)
ax[1].set_ylim([1e-8,1e2]);ax[1].set_xlabel('f [Hz]');ax[1].set_ylabel('$G_{xx} [V^2 Hz^{-1}]$')

ax[0].set_title('Filtered and Unfiltered Signal')
ax[1].set_title('Power Spectrum wi/wo Filtering')
ax[1].text(B,1,' B=0.5 Hz');plt.axvline(x=B,linestyle='--',color='k')
plt.tight_layout()
```

## Q2: Cubic Spline

Using the data below...

```python
t = np.array([0.,2.5,2.7,3.4,4.2,6.5,9.0])
x = np.cos(t*2*pi/3)
```

**1** Perform a cubic spline interpolation onto an even grid from 0 to 9.0 by 
>  1. creating the matrix from the notes
>  2. inverting and solving for the cubic co-efficients
>  3. using those polynomials to evaluate the interpolated data

 Compare with using `scipy.interpolate.interp1d`.
 Comment on any similarities or differences between the two results.</div>

```python
n = len(t)-1

delta = np.zeros(n)
for i in range(0, n):
    delta[i] = t[i+1]-t[i]  
    
alpha = np.zeros(n+1)
for i in range(1, n):
    alpha[i] = 3.*((x[i+1]-x[i])/delta[i] - (x[i] - x[i-1])/delta[i-1])

k = np.zeros((n+1, n+1))
k[0,0] = 1
k[n,n] = 1
 
for i in range(1, n):
    k[i, i-1] = delta[i-1]
    k[i, i] = 2*(delta[i] + delta[i-1])
    k[i, i+1] = delta[i] 

tt = np.linspace(0, 9, 100)
    
c = np.linalg.tensorsolve(k, alpha)
c = np.insert(c, 0, 0.)

b = np.zeros(n+1)
d = np.zeros(n+1)
for i in range(n):
    d[i] = (c[i+1] - c[i])/(3*delta[i])
    b[i] = (x[i+1] - x[i] - c[i]*delta[i]**2 - d[i]*delta[i]**3)/delta[i] 
    b[i+1] = b[i] + 2*c[i]*delta[i] + 3*d[i]*delta[i]**2

d[n] = -1*c[n]/(3*delta[n-1])

def q(s):
    i = np.searchsorted(t,s,side='right')-1
    ss = s-t[i]
    return (x[i] + b[i]*ss + c[i]*ss**2 + d[i]*ss**3) 
    
y = interp1d(t, x, kind='cubic')
xx = np.cos(tt*2*pi/3)

fig, ax=subplots(1,1,figsize=(6,6))
ax.plot(t, x, label='Orig')
ax.plot(tt, xx,'--',color='k', label= 'x')
ax.plot(tt, q(tt), label='Mine')
ax.plot(tt, y(tt), label='Scipy')
ax.set_title('Cubic Spline Interpolation')
plt.legend(loc='best')
plt.tight_layout()
```
