## Q1: Distribution of power spectral estimates

<div style='background:#F0F0F0'>**1** You **may** use `matplotlib.mlab.psd` for the following question, but you had best use it correctly!

Using normally distributed random noise time series of length $N=2048$, show using a Monte Carlo analysis and by comparison to the theoretical probability distribution functions that:  
</div>

   1. The raw spectral estimate is indeed distributed as $\chi^2_2$.
   2. Show that block averaging with no overlap is distributed as $\chi^2_{2N_{blocks}}$.
   3. Show that block averaging with 50% overlap Hanning windows is distributed as $\chi^2_{2N_{blocks}}$.

<div style='background:#F0F0F0'>You can choose your block length, but making it an integer divisor of 2048 will make your life a lot easier.  Show that as $N_{blocks}$ is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter).</div>


<div style='background:#F0F0F0'>HINT: for this time series the individual frequency estimates are indipendent samples of the distribution, so you can use them in compiling your distributions.</div>

<div style='background:#F0F0F0'>HINT: To get the pdf of the $\chi^2_\nu$ distribution correct, you need to "scale" by $\nu^{-1}$ where $\nu$ are the degrees of freedom.</div>

<div style='background:#F0F0F0'>HINT: for good presentation, make your histograms have the same bin sizes, and compare the distributions for all three cases on the same plot.</div>

```python
n = 2048
MC = 200

x = np.empty(1)

for i in range(0, MC):
    noise = np.random.randn(n)
    px, f = mlab.psd(noise, n-1, 1)
    x = np.append(x, px)

    
dfx = 2  
fig, ax=plt.subplots(1,1)
binx = ax.hist(x, 100, normed=True, histtype='step', label= 'PSD')
ax.plot(binx[1],stats.chi2.pdf(binx[1], dfx), color='r', label='$\chi^2_2$')
ax.set_title('Comparison of $\chi^2_2$ Distribution to PSD of Random Noise')
ax.set_ylabel('Frequency')
ax.set_xlabel('Spectral Density')
ax.legend(loc=1)
plt.tight_layout()
```

```python
block = 256
y = np.empty(1)

for i in range(0, MC):
    noise = np.random.randn(n)
    py, f = mlab.psd(noise, block-1, 1)
    y = np.append(y, py/2)

N = n/block
dfy = 2*N
banana = dfy**-1
fig, ay=plt.subplots(1,1)
biny = ay.hist(y, 100, normed=1, histtype='step', label='PSD')
ay.plot(biny[1],stats.chi2.pdf(biny[1], dfy, scale=banana), color='r', label='$\chi^2_{2N}$')
ay.set_title('Comparison of $\chi^2_{2N}$ Distribution to PSD of Random Noise')
ay.set_ylabel('Frequency')
ay.set_xlabel('Spectral Density')
ay.legend(loc=1)
plt.tight_layout()
```

**Comment:**
> After some work sorting out the appropriate scaling, the $\chi^2_{2N}$ distribution matches our random noise histogram with 200 Monte Carlo trials.

```python
block = 128
z = np.empty(1)

for i in range(0, MC):
    noise = np.random.randn(n)
    pz, f = mlab.psd(noise, block-1, 1, noverlap=block/2)
    z = np.append(z, pz[1:-2]/2)
    
N = 2*n/block
dfz = 2*N   
banana = dfz**-1
dft = 18/11*N
apple = dft**-1
fig, az=plt.subplots(1,1)
binz = az.hist(z, 100, normed=True, histtype='step', label='PSD')
az.plot(binz[1],stats.chi2.pdf(binz[1], dfz, scale=banana), color='r',label='$\chi^2_{2N}$')
az.plot(binz[1],stats.chi2.pdf(binz[1], dft, scale=apple), color='g',label='$\chi^2_{18N/11}$')
az.set_title('Comparison of $\chi^2$ Distributions to PSD with Overlap')
az.set_ylabel('Frequency')
az.set_xlabel('Spectral Density')
az.legend(loc=1)
plt.tight_layout()
```

## Q2: Spectra of a "red-noise" signal

<div style='background:#F0F0F0'> **1** Load in the Deep Cove hourly data, and compute the power spectra using some reasonable value for $N_{FFT}$, and comment on the effect of applying the Hanning window to not applying it to the spectral leakage.</div>

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
dc=dc[isfinite(dc)]
```

```python
blk = 451

dno, fno = mlab.psd(dc, NFFT=blk, Fs=1)
do, fo = mlab.psd(dc, NFFT=blk, Fs=1, noverlap=blk/2)

fig, ay=plt.subplots(1,1)
ay.loglog(fno[1:-2], dno[1:-2], label='No Overlap', color= 'black')
ay.loglog(fo[1:-2], do[1:-2], label='50% Overlap', color='r', alpha=0.6)
ay.set_title('NFFT: With and Without Overlap')
ay.set_xlabel('Frequency')
ay.set_ylabel('Spectral Density')
ay.legend(loc=1)
plt.tight_layout()
```

**Comment:**

>The overlapped and discrete windows show similar results. Close inspection reveals less variation in the data set with overlap, specifically in the regions of higher frequency (smoother near the bottom right). 

<div style='background:#F0F0F0'> **2** Now do the same thing to the integral of the Deep Cove data with time: $$y(t)=\int_o^t x(t')\ \mathrm{d}t'$$  This is obviously a silly thing to do, but compare the spectral and comment on the difference between the Hanning window and the non-Hanning windowed data.  </div>

```python
dcy = np.cumsum(dc)

fig, by=plt.subplots(1,1)
dnw, fnw = mlab.psd(dcy, NFFT=blk, Fs=1, window=mlab.window_none)
dw, fw = mlab.psd(dcy, NFFT=blk, Fs=1, window=mlab.window_hanning)
by.loglog(fnw[1:-2], dnw[1:-2], label='No Window')
by.loglog(fw[1:-2], dw[1:-2], label='Window')
by.set_title('NFFT: With and Without Hanning Window')
by.set_xlabel('Frequency')
by.set_ylabel('Spectral Density')
by.legend(loc=1)
plt.tight_layout()
```

**Comment: **

>Without the Hanning window, analysis of the power spectrum for the integral of the data is pretty much a straight line (all variation in the data has been eliminated). Adding the hanning window shows where spikes occur in the data while still allowing one to see the general trend.
